## 1. Composition(has-a)

Composition is a process of creating an object of one class inside another class or a class refers to the members of another class.

In composition the life time contained object is until container object is exists.

If container object is destroyed, contained object also destroyed.

example:

1. Book has a pages.

2. Car has a engine.

3. person has a address.

```python
class Engine:
    def start(self):
        print("Engine starts")
    def stop(self):
        print("Engine stop")
class Car:
    def __init__(self):
        self.e=Engine()
    def start(self):
        self.e.start()
    def stop(self):
        self.e.stop()
audi=Car()
audi.start()
audi.stop()
```

## Inheritance

Inheritance is a process of acquiring the properties & behaviour of one class inside another class.

Inheritance allows developing a news class or datatype using existing class or datatype.

Relationship(IS-A) is for related classes but not for unrelated classes.

Advantages of inheritance is reusability.

1. variables

2. Methods

`PARENT CLASS` can called as Super class,Parent class,Base class,Generalized class.

`CHILD CLASS` can called as Sub class,Derived class,Child class,Specialized class.

The class having common properties and behaviour is called Generalized class.

The types of inheritance:

1. Single level inheritance

2. Multi level inheritance

3. Multiple inheritance

4. Hybrid inheritance

5. Hierarchicall inheritance

### Single level inheritance

!

```python
class Parent:
    def show(self):
        print("This is the Parent class.")

class Child(Parent):  # Single inheritance
    def display(self):
        print("This is the Child class.")

obj = Child()
obj.show()    # Inherited method
obj.display() # Child's method
```

### 2. Multi Inheritance
Definition: A child class inherits from a parent class, and another child class inherits from that child class (forming a chain).

!

```python
class Father:
    def traits(self):
        print("Father's traits.")

class Mother:
    def characteristics(self):
        print("Mother's characteristics.")

class Child(Father, Mother):  # Multiple inheritance
    def own(self):
        print("Child's unique attributes.")

obj = Child()
obj.traits()           # Inherited from Father
obj.characteristics()  # Inherited from Mother
obj.own()              # Child's method
```

### 3. Multiple Inheritance
Definition: A child class inherits from more than one parent class.

!

### 4. Hierarchical Inheritance
Definition: Multiple child classes inherit from a single parent class.

!

```python
class Parent:
    def show(self):
        print("Parent's property.")

class Child1(Parent):
    def show_child1(self):
        print("Child1's property.")

class Child2(Parent):  # Hierarchical inheritance
    def show_child2(self):
        print("Child2's property.")

obj1 = Child1()
obj2 = Child2()

obj1.show()       # Inherited from Parent
obj1.show_child1()

obj2.show()       # Inherited from Parent
obj2.show_child2()
```

### 5. Hybrid Inheritance
Definition: A combination of two or more types of inheritance. It often involves multiple inheritance and hierarchical inheritance.

!

```python
class Base:
    def show_base(self):
        print("Base class.")

class Derived1(Base):
    def show_derived1(self):
        print("Derived1 class.")

class Derived2(Base):
    def show_derived2(self):
        print("Derived2 class.")

class Hybrid(Derived1, Derived2):  # Combination
    def show_hybrid(self):
        print("Hybrid class.")

obj = Hybrid()
obj.show_base()       # Inherited from Base
obj.show_derived1()   # Inherited from Derived1
obj.show_derived2()   # Inherited from Derived2
obj.show_hybrid()     # Hybrid's method
```

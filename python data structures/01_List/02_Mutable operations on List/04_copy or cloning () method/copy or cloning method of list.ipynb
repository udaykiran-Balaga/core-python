# Copy or cloning

python allows creating copy of collections.

It various methods or types of copies

`1. alias copy`

`2. shallow copy`

`3. deep copy`

### ALIAS COPY

Assigning address of one list to another is called alias copy.



In alias copy, same list object is assigned to multiple variables, In this approach new list object is not created.


```python
list1=list(range(10,60,10))
```

```python
print(list1)
```

```python
list2=list1
```

```python
print("ID of list1",id(list1))
print("ID of list2",id(list2))
```

see here two lists having same address, it means that its not created a new list from list1.

If we manipulate the values/items in the list2 or list1 then it effects on the another list

```python
list1.insert(0,100)
print(list1)
```

```python
print(list2)
```

see here two list at `index 0` insets with `100` value

```python
list2[0]=200
print(list2)
print(list1)
```

```python
del list1[-5]
print(list1)
print(list2)
```

### Shallow copy or Shallow Cloning

List provides a method called "copy".

This creates shallow copy of list.

In shallow copy a new list object is created by copying address of objects found in old list.



In shallow copy, if the objects exists inside list are mutable, the changes done in one list reflect to another list.

```python
list1 = [10,20,[30,40]]
list2=list1.copy()
print(id(list1))
print(id(list2))
```

```python
print(list1)
print(list2)
```

```python
list1[2][0]=99
print(list1)
```

```python
print(list2)
```

```python
list1.insert(10,100)
print(list1)
```

```python
print(list2)
```

```python
del list1[0]
print(list1)
```

```python
print(list2)
```

### Deep copy

In deep copy, a new list object created by copying objects found inside old list.

This deep copy is done using deep copy function available in copy module.



```python
list1=[10,20,[30,40]]
print(list1)
```

```python
import copy
list2=copy.deepcopy(list1)
print(list2)
```

```python
list1[2][0]=99
print(list1)
```

```python
print(list2)
```

```python
list2[2][0]=88
print(list2)
```

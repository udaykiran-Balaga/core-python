# Dictionary (mapping)

Dictionary mutable collection, after creating dictionary changes can be done.

Dictionary is mapping collection, wher data is orgainzed as key and value pair.

Each value in dictionary is identified with key.

one key is mapped with one or more than one value.

Dictionary does not allows duplicate keys but allowed duplicate values.

Dictionary keys are hashable objects/immutable objects.

In application developing to orgainze data as key and value pair use dictionary.

## How to create dictionary?

1. Empty dictionary is created using empty curly braces {}

```python
d1= {}
print(d1,type(d1))
```

2. create a dictionary with items, where each item consist of key and value with is separated using `:`

**Syntax: {key:value,key:value,key:value,….}**

```python
d2={1:10,2:20,3:30,4:40,5:50}
print(type(d2),d2)
```

```python
d3={1:10,1:20,1:30,1:40}
print(d3)
```

3. Creating dictionary using existing iterables/collections

**Syntax2: dict(iterable)  : convert existing iteable into dictionary**

```python
list1=[(1,10),(2,20),(3,30),(4,40),(5,50)]
d3=dict(list1)
print(d3)
```

```python
list2=[10,20,30,40,50]
e=enumerate(list2)
d4=dict(e)
print(d4)
```

```python
list1=[1,2,3,4,5]
list2=[10,20,30,40,50]
list3=[(list1[i],list2[i]) for i in range(5)]
print(list3)
```

```python
d5=dict(list3)
print(d5)
```

**zip(iterables,strict-False)**

iterate over several iterables in parallel, producing tuples with an item from each one.

**More formally**: `zip()` return an iterator of tuples, where the i-th tuple contains the i-th element from each of the argument iterables.

```python
d6=dict(zip(range(1,6),range(10,60,10)))
print(d6)
```

```python
d7=dict(zip("ABCD",range(65,69)))
print(d7)
```

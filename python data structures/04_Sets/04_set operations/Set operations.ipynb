# Set Operations

Advantage of using sets is performing mathematical set operations.

`union`, `intersection`, `difference`....

**union(others)**:



```python
a={1,2,3}
b={4,5,6}
c=a.union(b)
print(a,b,c,sep='\n')
```

```python
d=[6,7,8]
e=a.union(d)
print(e)
```

```python
# example

a={1,2,3}
b={4,5,6}
c={8,9,0}
d=a|b|c
print(d)
```

**intersection(others)**:

Return a new set with elements common to the set and all others.



```python
A={1,2,3,4,5}
B={1,2,3,6,7}
C=A&B
print(A,B,C,sep='\n')
```

**difference(others)**

Return a new set with elements in the set that are not in the other.



```python
A={1,2,3,4,5}
B={1,2,3,6,7}
C=A.difference(B)
print(A,B,C,sep='\n')
```

**symmetric difference(others)**

Return a new set with elements in either the set or other but not both
")


```python
A={1,2,3,4,5}
B={1,2,3,6,7}
C=A.symmetric_difference(B)
print(A,B,C,sep="\n")
```

```python
A={10,20,30,40,50}
B={10,20,30,60,70}
C=A^B
print(A,B,C,sep="\n")
```

**update(others)**

Update the set, adding elements from all others.


```python
A={1,2,3,4}
A.update({1,2,3,4,5,6,7})
print(A)
```

```python
A={1,2,3}
B={4,5,6}
A|=B
print(A)
```

**intersection_update(others)**

Update the set, keeping only elements found in it and all others.

```python
A={1,2,3,4,5}
B={1,2,6,7,8}
A.intersection_update(B)
print(A)
```

```python
A={10,20,30,40,50}
B={10,20,30,60,70}
A &= B
print(A)
```

**difference_update(others)**

Update the set,removing element found in others

```python
A={1,2,3,4,5}
B={1,2,3}
A.difference_update(B)
print(A)
```

**symmentric_difference_update(others)**

update the set, keeping only elements found in either set, but not in both.

```python
A={1,2,3,4}
B={1,2,5,6}
A^=B
print(A)
```

## Set examine methods

1. **isupperset(other)**

`set >= other`

Test whether every element in other is in the set.

2.	**issubset(other)**

`set <= other`
Test whether every element in the set is in other

3.	**isdisjoint(other)**

Return True if the set has no elements in common with other. Sets are disjoint if and only if their intersection is the empty set.
.


**isupperset(other)**

```python
A={1,2,3,4,5}
B={1,2,3,4,5}
A>=B
```

```python
A={1,2,3,4,6}
B={1,2,3,4,5}
A.issuperset(B)
```

**isubset(other)**

```python
A={1,2,3}
B={1,2,3,4,5}
A<=B
```

**isdisjoint(other)**

Return True if the set has no elements in common with other. Sets are disjoint if and only if their intersection is the empty set.

```python
A={1,2,3}
B={4,5,6}
A.isdisjoint(B)
```

```python
C={1,6,7}
A.isdisjoint(C)
```

# Bitwise operators

Bitwise operators are used to perform operation on binary data.

Bitwise operators are:

shift Operators 

a . >> (Right shift operator) 

b . << (Left shift operator)

1. Bitwise `&` (and) operator
2. Bitwise `!` (or) operator
3. Bitwise `^` (xor) operator
4. Bitwise `~` (not) operator

**NOTE**

bitwise oeprators are applied only one integer data types

# Right shift operator

This oeprator is used to `shift number` of bits towards `right side`.

By shifting number bits towards right side the value get `decremented` or `shifted bits are deleted`.

## Examples on Bitwise Operators

```python
s=0b1010

print(s)
```

```python
print(bin(s))
```

```python
b=s>>1
```

```python
print(b)
```

```python
print(bin(b))
```

```python
bin(25)
```

```python
bin(0o12)
```

```python
bin(0xb)
```

```python
n1=65
```

```python
n2=n1>>3
```

```python
print(n1,n2)
```

```python
print(bin(n1),bin(n2))
```

## Left Shift Operator

This operator is used to shift number of bits towards left side.

By shifting number of bits towards left side the value get incremented by adding those many nuber bits at right side.

```python
a=12
b=a<<1
print(a)
print(b)
```

```python
print(bin(a))
print(bin(b))
```

# Bitwise **`&`** oeprator

This bitwise operator is used to apply `and` gates

Truth table of **`and`** **(&)** operator

Here two operands are needs operand1-`A` and operand2-`B` --> **`A&B=Q`**

1 --> True

0 --> False

**False+False=False**

**False+True=False**

**True+False=False**

**True+True=True**

```python
A=0b1
B=0b1
A&B
```

##  Bitwise `(|)` operator

This bitwise oeprator is used to apply on operators `or` gates

Truth table of **`or`** **(|)** operator

1 --> True

0 --> False

**False+False=False**

**False+True=True**

**True+False=True**

**True+True=True**

## Bitwise (^) XOR operator

```python
A=0b1
B=0b0
A|B
```

```python
X=0b1010
Y=0b1110
Z=X|Y
print(X,Y,Z)
```

This bitwise operator is used to apply `XOR` gate

```python
A=2
B=3
C=A^B
```

```python
print(A,B,C)
print(bin(A),bin(B),bin(C))
```

## Bitwise not operator

This operator is used inverse bits

It is a unary operator, it required one operand

Truth table of not operator

```python
a=0b1
b=~a
print(bin(a),bin(b))
print(a,b)
```

```python
x=0b101
y=~x
print(bin(x),bin(y))
print(x,y)
```

```python
a=7
b=~a
print(a,b)
```

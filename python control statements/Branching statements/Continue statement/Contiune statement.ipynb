# continue

`continue` is a keyword is used inside `while` or `for` loop

This statement continues the execution of for or while loop.

syntax-1:

`while <condition>:`

    statement-1

    continue

syntax 2:

`for variable in iterable:`

    statement-1

    continue

```python
# Example:

for i in range(1,6):
    print("Python")
    continue
    print("Django")
```

```python
for n in range(1,11):
    if n<=5:
        print(n)
        continue
    break
```

```python
i = 0
while i < 9:
  i += 1
  if i == 3:
    continue
  print(i)
```

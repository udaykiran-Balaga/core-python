# While loop

Looping control statements are used to repeat one or more than one statement number of time or until given condition.

Python support two types of looping control statements

`1. while`

`2. for`

**while loop**

whilel oop is used to execute or repeat block of statemebts until given condition or Booean expression

**syntax-1**

`while <condiition>:`

    statement-1

    statement-2

statement-3

here statement-1,statement-2 are repeated until given condition is True, if condition is Flase, it terminates loop adn continue with statement-3

**syntax**

`while <condition>:`

    statement-1

    statement-2

`else:`

    statement-3

    statement-4

statement-5

while loop required 3 statements

`1. initialization statement`

`2. condiiton`

`3. update statement`

`initalization statement` defines the initial value of condition.

`condition or Boolean expression` which defines how may times whole loop has to repeated

`update statement` which updates condition

```python
# example
        
n=1              # initialization statement
while n<=5:      # conditiona
    print("Hello")
    n+=1         # update statement
```

```python
# write a prohram to print numbers from 1 to 10

num=1
while num<=10:
    print(num,end=' ')
    num+=1
```

```python
# write a program to find length of number

num = int(input("Enter a number "))
c=0

while num>0:
    c=c+1
    num=num//10
print(f'Length of number or count of digits {c}')
```
